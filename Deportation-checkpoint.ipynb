```python
df['Month-Year'] = pd.to_datetime(df['Month-Year'], format='%b %Y')
df['Month-Year']
```

```python
#quero criar uma coluna com os anos civis, para saber qual foi o que teve mais removals.Uso a função do dt do pandas para achar anos.
```

```python
df['Civil_Year'] = df['Month-Year'].dt.year
df['Civil_Year']
```

```python
#mas eu quero saber não o número de colunas que os anos aparecem, mas quantas remoções teve em cada ano.
```

```python
year_real = df.groupby('Civil_Year')['Removals'].sum().sort_values(ascending=True).reset_index()
year_real
```

```python
year_real.to_csv('year_real.csv', index=False)
```

```python
#qual a porcentagem de aumento?
```

```python
inicio = year_real['Removals'].iloc[0]   # 2020
fim = year_real['Removals'].iloc[-1]    # 2024
```

```python
aumento_total = ((fim - inicio) / inicio) * 100
aumento_total
```

```python
(f"{aumento_total:.1f}%")
```

```python
ano_antes = year_real['Removals'].iloc[-2]
```

```python
aumento_final = ((fim - ano_antes) / ano_antes) * 100
aumento_final
```

```python
vezes = fim / inicio
vezes
```

```python
(f"{vezes:.1f}")
```
